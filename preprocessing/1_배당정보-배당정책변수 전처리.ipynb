## 배당정책
- 주당 배당금 기준으로 오르면 (1)/ 내리거나 동일하면 (0)

```python
#이진분류 함수를 생성
def bina(a, b):
  if a <= b:
    return 0
  elif a > b:
    return 1
```

```python
bina(2,1)
```

```python
df_3['배당정책(Y)'] = df_3.apply(lambda x: bina(x['2021 주당배당금'], x['2020 주당배당금']), axis = 1)
```

## 배당률
- = (주당배당금 / 액면가) * 100
- __<u>astype함수로 int나 float로 바꾸려고 하면 오류가 발생함</u>__
  - 문제 해결
    - __df[df['-'].str.contains('찾을문자열')]__을 사용하여 문자열 확인
    - __df['-'].str.replace(',', '').astype('int64')__ 사용하여 타입 변경하기
- __str.replace__함수 알아보기


```python
type(df['주당배당금'][3])
```

```python
#현금배당액면가 int로 타입변경
df['현금배당 액면가'] = df['현금배당 액면가'].str.replace(',','').astype('int64')
```

```python
df['현금배당 액면가'] = df['현금배당 액면가'].astype('float64')
```

```python
type(df['현금배당 액면가'][3])
```

```python
#배당률 구하는 함수를 정의
def divi_yield(a, b):
  try:
      divi = (a / b) * 100
  except ZeroDivisionError:   #ZeroDivisionError를 무시하기 위해 try ~ except구문 사용
    divi = 0
  return divi
```

```python
#df에 배당률을 추가
df['배당률'] = df.apply(lambda x: divi_yield(x['주당배당금'], x['현금배당 액면가']), axis = 1)
```

#### 배당률의 한계??
- __상장기업의 경우__ 액면가 보다는 __주식시가(시세)에 의해 기업의 주식이 거래__된다는 점에서 <br><u>배당률만으로 배당금이 현재 시장가치 대비 어느 정도 차지하는지 알기 어려움</u>
<br>
- 기업의 배당관련투자분석 시 <br>배당률 지표만 단독적으로 활용하기 보다는 <u>__시가배당률__과 __배당수익률__ 지표를 보조적으로 활용함!!</u>
 - 시가배당률과 배당수익률
    - 시가배당률과 배당수익률은 주식의 시장가치인 <u>주가(주식시세)를 기준으로 계산</u>한다는 점에서 배당률과 차이가 있음
