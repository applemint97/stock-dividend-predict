## 당기순이익(손실)
- __항목코드__로 찾기!!
- 항목코드 : 'ifrs-full_ProfitLoss'

```python
#df_1=df[(df['항목명']=='당기순이익')|(df['항목명']=='당기순이익(손실)')|(df['항목명']=='당기순손실')]
#|(df['항목코드']=='ifrs-full_ProfitLossFromContinuingOperations')제외
```

```python
df_1 = df[(df['항목코드']=='ifrs-full_ProfitLoss')|
          (df['항목명']=='당기순손실')|(df['항목명']=='당기순이익')|(df['항목명']=='당기순이익(손실)')|(df['항목명']=='당기순이익(순손실)')|
          (df['항목명']=='VI. 당기순이익(손실)')|(df['항목명']=='IX. 당기순이익(손실)')|(df['항목명']=='당기순이익(손 실)')|
          (df['항목명']=='Ⅶ. 당기순이익(손실)')|(df['항목명']=='Ⅷ.당기순이익(손실)')|(df['항목명']=='XIV. 당기순손실')|(df['항목명']=='Ⅷ. 당기순이익')]
```

```python
df_1.columns
```

```python
df_1.columns = ['종목코드', '회사명', '시장구분', '업종', '업종명', '결산월', '결산기준일', '보고서종류', '통화',
       '항목코드', '항목명', '당기순이익(손실)_당기', '당기순이익(손실)_전기', '당기순이익(손실)_전전기']
```

```python
#불필요한 열 제거
df_1 = df_1[['회사명', '당기순이익(손실)_당기', '당기순이익(손실)_전기', '당기순이익(손실)_전전기']]
```

```python
df_1
```

# 주당순이익
- 항목코드 : ifrs-full_BasicEarningsLossPerShare
  - 주당순이익은 항목코드가 다른 회사들도 있어 데이터셋 완성후 csv파일과 비교해가며 결측치를 채워줘야 함

```python
df_2 =  df[(df['항목코드']=='ifrs-full_BasicEarningsLossPerShare')|(df['항목명']=='보통주 기본주당순이익')|(df['항목명']=='      보통주 기본주당순이익')|
           (df['항목명']=='   보통주의 기본주당순이익')|(df['항목명']=='   기본(희석)주당순이익(손실)')|(df['항목명']=='   기본및희석주당순이익')|
           (df['항목명']=='   기본주당순이익')|(df['항목명']=='   보통주 기본주당순이익')|(df['항목명']=='   보통주당순이익')|(df['항목명']=='   기본주당순이익(손실)')|
           (df['항목명']=='      보통주 기본주당순이익')|(df['항목명']=='   기본주당순이익(보통주)')|(df['항목명']=='   기본 및 희석주당순이익(손실)')|(df['항목명']=='   기본및희석주당순이익')|
           (df['항목명']=='   보통주 기본 및 희석주당순이익(손실)')]
```

```python
df_2.shape
```

```python
df_2.columns = ['종목코드', '회사명', '시장구분', '업종', '업종명', '결산월', '결산기준일', '보고서종류', '통화',
       '항목코드', '항목명', '기본주당순이익(손실)_당기', '기본주당순이익(손실)_전기', '기본주당순이익(손실)_전전기']
```

```python
#불필요한 열 제거
df_2 = df_2[['회사명', '기본주당순이익(손실)_당기', '기본주당순이익(손실)_전기', '기본주당순이익(손실)_전전기']]
```

```python
df_2
```

## 영업이익
- 항목코드 : dart_OperatingIncomeLoss

```python
#영업이익 행만 추출
df_3 = df[(df['항목코드']=='dart_OperatingIncomeLoss')|(df['항목명']=='영업이익(손실)')|(df['항목명']=='영업이익')]


#영업이익 열이름 바꾸기
df_3.columns = ['종목코드', '회사명', '시장구분', '업종', '업종명', '결산월', '결산기준일', '보고서종류', '통화',
       '항목코드', '항목명', '영업이익(손실)_당기', '영업이익(손실)_전기', '영업이익(손실)_전전기']


#불필요한 열 제거
df_3 = df_3[['회사명', '영업이익(손실)_당기', '영업이익(손실)_전기', '영업이익(손실)_전전기']]
```

```python
df_3
```

## 매출액
- 항목코드 : ifrs-full_Revenue

```python
#매출액 행만 추출
df_4 = df[(df['항목코드']=='ifrs-full_Revenue')|(df['항목명']=='매출액')|(df['항목명']=='수익(매출액)')]


#영업이익 열이름 바꾸기
df_4.columns = ['종목코드', '회사명', '시장구분', '업종', '업종명', '결산월', '결산기준일', '보고서종류', '통화',
       '항목코드', '항목명', '매출액_당기', '매출액_전기', '매출액_전전기']


#불필요한 열 제거
df_4 = df_4[['회사명', '매출액_당기', '매출액_전기', '매출액_전전기']]
```

```python
df_4
```
