# 당기순손실여부(X5)
- 2년간 당기순손실 발생 (1), 아니면 (0)
- 당기순손실이 2년 중 한번이라도 발생하면 1

```python
#필요한 열 추출
df_1 = df[['회사명', '당기순이익(손실)_당기', '당기순이익(손실)_전기']]
```

```python
df_1
```

```python
#정수형 확인
df_1.info()
```

```python
#당기순손실여부 (0/1)함수 생성 a=2021, b=2020
def mi(a, b):
  if a < 0 or b < 0:
    return 1
  elif a > 0 and b > 0:
    return 0
```

```python
df_1['당기순손실여부(X5)'] = df_1.apply(lambda x: mi(x['당기순이익(손실)_당기'], x['당기순이익(손실)_전기']), axis=1)
```

```python
df_1 = df_1[['회사명', '당기순손실여부(X5)']]
```

```python
df_1
```

```python
#결측치 확인
df_1.isna().sum()
```

## 유동비율(X7)
- (유동자산/유동부채) * 100

```python
#유동자산과 유동부채 열추출
df_3 = df[['회사명', '유동자산_당기', '유동자산_전기', '유동부채_당기', '유동부채_전기']]
```

```python
df_3.info()
```

```python
df_3
```

```python
#유동비율 함수 생성 a = 유동자산, b = 유동부채
def fl(a, b):
  try:
      flu = (a / b) * 100
  except ZeroDivisionError:   #ZeroDivisionError를 무시하기 위해 try ~ except구문 사용
    flu = 0
  return round(flu, 2)  #뒤에 두자리수로 반올림
```

```python
#유동비율 당기 열 추가
df_3['유동비율_당기(X7)']= df_3.apply(lambda x: fl(x['유동자산_당기'], x['유동부채_당기']), axis=1)
```

```python
#유동비율 전기 열 추가
df_3['유동비율_전기(X7)']= df_3.apply(lambda x: fl(x['유동자산_전기'], x['유동부채_전기']), axis=1)
```

```python
df_3 = df_3[['회사명', '유동비율_당기(X7)', '유동비율_전기(X7)']]
```

```python
df_3
```

## 부채비율(X8)
- (총부채(부채총계) / 자기자본(자본총계)) * 100

```python
df_4 = df[['회사명', '부채총계(총부채)_당기', '부채총계(총부채)_전기', '자본총계(자기자본)_당기', '자본총계(자기자본)_전기']]
```

```python
df_4
```

```python
#부채비율 함수 생성 a = 부채총계, b = 자본총계
def dep(a, b):
  try:
      dept = (a / b) * 100
  except ZeroDivisionError:   #ZeroDivisionError를 무시하기 위해 try ~ except구문 사용
    dept = 0
  return round(dept, 2)  #뒤에 두자리수로 반올림
```

```python
#부채비율_당기 열 추가
df_4['부채비율_당기(X8)'] = df_4.apply(lambda x: dep(x['부채총계(총부채)_당기'], x['자본총계(자기자본)_당기']), axis=1)

#부채비율_전기 열 추가
df_4['부채비율_전기(X8)'] = df_4.apply(lambda x: dep(x['부채총계(총부채)_전기'], x['자본총계(자기자본)_전기']), axis=1)
```

```python
#부채비율
df_4 = df_4[['회사명', '부채비율_당기(X8)', '부채비율_전기(X8)']]
```

## 총자산증가율(X10)
- [(기말총자산(자산총계) / 기초총자산) / 기초총자산] * 100

```python
df_6 = df[['회사명', '자산총계_당기', '자산총계_전기', '자산총계_전전기']]
```

```python
df_6
```

```python
#총자산증가율 함수 생성 a = 금년도 총자산, b = 전년도 총자산
 
def asset(a, b):
  try:
      a = ((a - b) / b) * 100
  except ZeroDivisionError:   #ZeroDivisionError를 무시하기 위해 try ~ except구문 사용
    a = 0
  return round(a, 2)  #뒤에 두자리수로 반올림
```

```python
#총자산증가율_당기 열 추가
df_6['총자산증가율_당기(X10)'] = df_6.apply(lambda x: asset(x['자산총계_당기'], x['자산총계_전기']), axis=1)

#총자산증가율_전기 열 추가
df_6['총자산증가율_전기(X10)'] = df_6.apply(lambda x: asset(x['자산총계_전기'], x['자산총계_전전기']), axis=1)
```

```python
#총자산증가율
df_6 = df_6[['회사명', '총자산증가율_당기(X10)', '총자산증가율_전기(X10)']]
```
