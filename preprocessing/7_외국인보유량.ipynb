```python
import pandas as pd

df1 = pd.read_csv("./2021_외국인지분율.csv", encoding = 'euc-kr')
df2 = pd.read_csv("./2020_외국인지분율.csv", encoding = 'euc-kr')
df3 = pd.read_csv("./2019_외국인지분율.csv", encoding = 'euc-kr')
```

```python
df = pd.merge(df1, df2, how = 'outer', on = "종목코드")
df = pd.merge(df, df3, how = 'outer', on = "종목코드")
df
```

```python
df = df.dropna(axis = 0)
df
```

```python
df.columns
```

```python
df.columns = ['종목코드', '종목명_x', '2021_외국인지분율', '종목명_y', '2020_외국인지분율', '회사명', '2019_외국인지분율']
df = df[['종목코드', '회사명', '2021_외국인지분율', '2020_외국인지분율', '2019_외국인지분율']]
df
```

```python
외국인지분율변동_2021 = []

for i in range(len(df['2021_외국인지분율'])):
    if df['2021_외국인지분율'].iloc[i] < df['2020_외국인지분율'].iloc[i]:
        외국인지분율변동_2021.append(1)
    else:
        외국인지분율변동_2021.append(0)
        
외국인지분율변동_2021
```

```python
외국인지분율변동_2020 = []

for i in range(len(df['2020_외국인지분율'])):
    if df['2020_외국인지분율'].iloc[i] < df['2019_외국인지분율'].iloc[i]:
        외국인지분율변동_2020.append(1)
    else:
        외국인지분율변동_2020.append(0)
        
외국인지분율변동_2020
```

```python
df['외국인지분율변동_2021(X18)'] = 외국인지분율변동_2021
df['외국인지분율변동_2020(X18)'] = 외국인지분율변동_2020
```
